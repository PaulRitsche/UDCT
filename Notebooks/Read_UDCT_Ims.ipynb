```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
import random
```

## Segmented images

```python
# Open File containing images
f = h5py.File('first_test_model_gen_B.h5')

# Look at content
print(f.keys())
print(f['B'].keys())

# Extract the data
data = np.array(f['B/data'])
print(data.shape)

# Close the file (only works, because of np.array. Without this conversion, you cannot access 'data' anymore)
f.close()
```

```python
# Plot 5 randomly selected images
plt.figure(figsize=(25,25))

nums = random.sample(range(10), 5)
for i in range(len(nums)):
    plt.subplot(1,5,1+i)
    plt.imshow(data[nums[i],:,:,0]/(2**16-1.),cmap='gray')
    plt.axis('off')
plt.show()
```

```python
# Save the images
my_dpi = 120 # https://stackoverflow.com/questions/13714454/specifying-and-saving-a-figure-with-exact-size-in-pixels
save_folder = "Synthetic/"
for im in range(0, len(data)):
    fig = plt.figure(frameon=False)
    fig.set_size_inches(512/my_dpi, 512/my_dpi)
    plt.axis('off')
    plt.grid(False)
    ax = plt.Axes(fig, [0., 0., 1., 1.])
    ax.set_axis_off()
    fig.add_axes(ax)
    savename = save_folder + 'seg_' + str(im) + '.png'
    plt.imshow(data[im].squeeze(), cmap='gray')
    plt.savefig(savename, dpi=my_dpi)
    plt.close(fig)
```
